```python
plt.figure(figsize=(10, 5))

keys = list(caseloads2010.keys())
vals = [float(caseloads2010[k]) for k in keys]

rank = [int((max(vals)-elem)*len(vals)*0.75/(max(vals)+1)) for elem in vals] 
pal = sns.color_palette("YlGnBu",len(vals))

ax = sns.barplot(x=keys, y=vals, palette=np.array(pal[::-1])[rank])

ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

plt.yticks([0, 20, 40, 60, 80, 100])

plt.title("Case Load 2010")
plt.ylabel("Case Load")
plt.xlabel("States")

plt.tight_layout()
plt.savefig('caseLoad2010.png')

plt.show()
```
